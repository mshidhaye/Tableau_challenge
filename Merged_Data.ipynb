```python
import pandas as pd
```

```python
data_2019 = pd.read_csv('results_data/data_2019.csv')
data_2019
```

```python
data_2020 = pd.read_csv('results_data/data_2020.csv')
data_2020
```

```python
total_merged_data_df = pd.concat([data_2020,data_2019],ignore_index=True)
total_merged_data_df
```

```python
# Dictionary mapping values in the 'gender' column to new values
mapping = {0: 'Unknown', 1: 'Male', 2: 'Female'}

# Use the 'map' method to change the values in the 'gender' column
total_merged_data_df['gender'] = total_merged_data_df['gender'].map(mapping)

# Show DataFrame
total_merged_data_df
```

```python
total_merged_data_df.to_csv('results_data/total_merged_data.csv', index=False)
```
